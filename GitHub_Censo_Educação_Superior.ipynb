# **MEC e Inep divulgam dados do Censo da Educação Superior 2018**

Em 2018, 37.962 cursos de graduação foram ofertados em 2.537 instituições de educação superior (IES) no Brasil para um total de 8.450.755 estudantes matriculados. Os dados são do Censo da Educação Superior e foram divulgados em coletiva de imprensa com a presença dos funcionarios do Ministério da Educação (MEC) e do Instituto Nacional de Estudos e Estatísticas Educacionais Anísio Teixeira (Inep).

Segundo as estatísticas apresentadas, as 199 universidades existentes no país equivalem a 7,8% do total de IES, mas concentram 52,9% das matrículas em cursos de graduação.

## Cursos

Os cursos de bacharelado mantêm sua predominância na educação superior brasileira com uma participação de 58% das matrículas. Os cursos tecnológicos tiveram o maior crescimento (16,6%) entre os graus acadêmicos em 2018, quando comparado a 2017.

## Vagas

Em 2018, foram oferecidas mais de 13,5 milhões de vagas em cursos de graduação, sendo 72,9% vagas novas e 26,9%, vagas remanescentes. Das novas vagas oferecidas em 2018, 30,6% foram preenchidas, enquanto apenas 11,3% das vagas remanescentes foram ocupadas no mesmo período.

## Ingressantes

Em 2018, 3,4 milhões de alunos ingressaram em cursos de educação superior de graduação. Desse total, 83,1% em instituições privadas. O número de ingressantes teve um crescimento de 6,8% em relação a 2017. O aumento no número de ingressantes em 2018 foi sustentado pelo significativo aumento na nos cursos
na modalidade a distância, compensando a queda registrada nos cursos presenciais.

## Concluintes

Em 2018, mais de um 1,2 milhão de estudantes concluíram a educação superior.O número de concluintes em cursos de graduação presenciais teve aumento de 4,5% em relação a 2017. A modalidade a distância aumentou 8,6% no mesmo período.

## Censo da Educação Superior

O Censo da Educação Superior, realizado anualmente pelo Inep, constitui-se importante instrumento de obtenção de dados para a geração de informações que subsidiam a formulação, o monitoramento e a avaliação das políticas públicas, além de ser elemento importante para elaboração de estudos e pesquisas sobre o setor. O Censo coleta informações sobre as Instituições de Educação Superior (IES), os cursos de graduação e sequenciais de formação específica e sobre os discentes e docentes vinculados a esses cursos.

Os resultados coletados subsidiam o Sistema Nacional de Avaliação da Educação Superior (Sinaes), seja no cálculo dos indicadores de Conceito Preliminar de Curso (CPC) e do Índice Geral de Cursos (IGC), seja no fornecimento de informações, como número de matrículas, de ingressos, de concluintes, entre outras. As estatísticas possibilitam ainda, através da justaposição de informações de diferentes edições da pesquisa, a análise da trajetória dos estudantes a partir de seu ingresso em determinado curso de graduação, e, consequentemente, a geração de indicadores de acompanhamento e de fluxo na educação superior.

### **SQL**: convertindo os arquivos *.csv em arquivos *.db 

```python
#função para convertir arquivos *.csv em *.db
def convert_csv_to_sql(path_to_data, db_name, chunksize):
    '''
    Função para conversão de arquivos csv em banco de dados sqlite3.
    
    INPUTS:
        path_to_data : str
            Caminho para os dados em formato csv a serem convertidos.
            
        db_name: str
            Nome do banco de dados a receber os dados
            
        chunksize : int
            Tamanho do lote de carregamento do arquivo csv.
    '''
    #criando conexão/banco de dados
    db = sqlite3.connect(db_name)

    #lendo todos os arquivos do caminho especificado em formato *.csv 
    files = glob.glob(f'{path_to_data}/*.CSV')
    
    #para cada um dos arquivos
    for file in files:

        #definindo o nome da tabela como sendo o nome do arquivo sem o *.csv
        #sim executando dum diretorio do PC: .split('\\')
        #sim executando dum diretorio do Google Drive: .split('/')
        table_name = file.split('/')[-1].split('.CSV')[0] 

        #carregando o arquivo indicado em chunks (lotes) de forma
        # a permitir operação com dataframes grandes demais
        df_chunks = pd.read_csv(file, sep='|', encoding='ISO-8859-1', chunksize=chunksize, low_memory=False)
        #alternativamente, usando encoding='latin1' obtem o mesmo resultado

        #para cada lote ("pedaço") do dataframe
        for df in df_chunks:

            #escrevendo o pedaço no banco
            df.to_sql(
                name=table_name, #definindo o nome da tabela a receber as informações (ou ser criada)
                index=False, #não escreve o índice do dataframe na tabela
                if_exists='append', #se a tabela já existir, adiciona as informações abaixo das existentes
                con=db #definindo em qual banco os dados serão inseridos
            )
```

A execução da função tem uma duração de uns 16 minutos. Vários dos arquivos *.csv carregados são grandes demais (e.g., > 2GB) com muitas linhas (e.g., 12043993 - ver abaixo).

```python
#utilizando a função para conversão dos *.csvs em banco de dados sqlite
convert_csv_to_sql('/content/gdrive/My Drive/Colab Notebooks/dados', 'desafio_1.db', chunksize=100000)
```

Os valores 1, 2 e 3 são Universidades Públicas, e os valores 4 e 5 são Universidades Privadas. Podemos utilizar as seguintes funções para agrupar essas categorias. 

```python
#criando categoria de Universidades Públicas e Privadas a partir de 'TP_CATEGORIA_ADMINISTRATIVA'
def classificar_publica(categoria_administrativa):
    if categoria_administrativa < 3:
        return 'Universidade Pública'  

def classificar_privada(categoria_administrativa):
    if categoria_administrativa > 3 < 5:
        return 'Universidade Privada'        

#agregando uma nova coluna 'CLF_PUBLICA_PRIVADA' com os valores trocados da coluna 'TP_CATEGORIA_ADMINISTRATIVA'
df_pub_pvd['CLF_PUBLICA'] = df_pub_pvd['TP_CATEGORIA_ADMINISTRATIVA'].apply(classificar_publica)
df_pub_pvd['CLF_PRIVADA'] = df_pub_pvd['TP_CATEGORIA_ADMINISTRATIVA'].apply(classificar_privada)

#tirando a coluna 'TP_CATEGORIA_ADMINISTRATIVA'
df_pub_pvd.drop('TP_CATEGORIA_ADMINISTRATIVA', axis='columns', inplace=True)
df_pub_pvd
```

```python
# SQL

query = '''
SELECT t2.NO_IES, COUNT(t1.ID_ALUNO) AS NUM_ALUNOS
FROM DM_ALUNO t1
INNER JOIN DM_IES t2
ON t1.CO_IES = t2.CO_IES
GROUP BY t2.NO_IES
ORDER BY NUM_ALUNOS DESC
'''

df_estudantes_univ = pd.read_sql_query(query, db)
df_estudantes_univ
```

```python
#selecionando as 30 maiores Universidades do Brasil
df_estudantes_univ_30 = df_estudantes_univ.sort_values(by='NUM_ALUNOS', ascending=False).iloc[:30]
#colocando todos os nomes das Universidades em letras maiúsculas
df_estudantes_univ_30['NO_IES']= df_estudantes_univ_30['NO_IES'].str.upper()
#selecionado a coluna 'NO_IES' como index
df_estudantes_univ_30 = df_estudantes_univ_30.set_index('NO_IES')
df_estudantes_univ_30
```

A presença de mulheres nos cursos de exatas pode ser descrita como uma distribuição binomial. Vamos então calcular a média e o desvio padrão desta distribuição. 

**Distribuição Binomial:**

p &rarr; probabilidade de sucesso (probabilidade de ser mulher)

$$p = \frac{n_{mulheres}}{n_{total}} $$

$$ \mu = n_{total} * p $$

$$ \sigma = \sqrt{n_{total} * p * (1-p)} $$

**Sources**: https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_binomial

```python
#cálculo de p
n_mulheres = df_estudantes_exatas_pd.loc[df_estudantes_exatas_pd['TP_SEXO'] == 1, 'ID_ALUNO'].count()
n_total = df_estudantes_exatas_pd['ID_ALUNO'].count()
p = n_mulheres / n_total

#cálculo da média
mu = n_total*p

#cálculo do desvio padrão
std = np.sqrt(n_total * p * (1-p))
```

```python
print(f'número de mulheres: {n_mulheres}','\n' f'número total: {n_total}','\n' f'mu: {mu}','\n' f'desvio padrão: {std}')
```

Agora, vamos fazer uma **distribuição normal** de uma população hipotética com a média $\mu$ e o desvio padrão $\sigma$ calculados para descrever a presença de mulheres nos cursos de exatas. Como queremos comparar a média de nossa amostra contra a média da população hipotética, podemos utilizar um teste t. Dessa maneira, podemos verificar se a média observada é estatisticamente menor que a média esperada caso tivessem o mesmo número de homens e mulheres nos cursos de exatas, ou seja, metade do número de estudantes.

Vamos utilizar um teste t com grau de significância $\alpha$ de 0.05, com as seguintes hipóteses:

$$ H_0 : \mu_{observada} = \frac{n_{total}}{2} $$

$$ H_1 : \mu_{observada} < \frac{n_{total}}{2}$$

Vale perceber que estamos fazendo um teste monocaudal, testando se o número de mulheres é menor do que o número de homens. Neste caso, o p-valor do teste monocaudal corresponde à $p_{bicaudal} / 2 $. Além disso, podemos olhar para o sinal do t-statistic para determinar se  média observada é maior ou menor que a média esperada.
- t_statistic < 0 e p-valor < $\alpha$ &rarr; A média observada é menor que a média esperada;
- t_statistic > 0 e p-valor < $\alpha$ &rarr; A média observada é maior que a média esperada;
- p-valor > $\alpha$ &rarr; A média observada e a média esperada não são estatisticamente diferentes;

**Sources**: https://pt.wikipedia.org/wiki/Teste_t_de_Student

```python
#definindo grau de significância
alpha = 0.05
#criando a distribuição normal hipotética de média mu e desvio padrão std
norm_dist_estudantes_exatas = stats.norm.rvs(loc=mu, scale=std, size=n_total)

#fazendo o teste de hipótese
t_statistic, p_value = stats.ttest_1samp(norm_dist_estudantes_exatas, int(n_total/2))

if p_value/2 >= alpha and t_statistic < 0:
    print(f'Com um p_valor de {round(p_value, 2)} não rejeitamos a hipótese nula, ou seja, não existem menos mulheres do que homens nos cursos de exatas!')
else:
    print(f'Com um p_valor de {round(p_value, 2)} rejeitamos a hipótese nula, ou seja, existem, de fato, menos mulheres do que homens nos cursos de exatas!')
```

A presença de mulheres nos cursos de exatas também pode ser avaliada com um **teste de independência do Qui-cuadrado**. Ele determina se existe uma relação estatisticamente significativa entre variáveis ​​categóricas, nosso caso os gêneros vs. os cursos de exatas. É outro teste de hipótese que responde à pergunta - os valores de uma variável categórica dependem do valor de outras variáveis ​​categóricas? Portanto:

**Hipótese nula (H0)**: não há relações entre as variáveis ​​categóricas. Se você souber o valor de uma variável, isso não o ajudará a prever o valor de outra variável. 

**Hipótese alternativa (H1)**: Existem relações entre as variáveis ​​categóricas. Saber o valor de uma variável ajuda a prever o valor de outra variável.

O teste de associação Qui-quadrado funciona comparando a distribuição que você observa com a distribuição que você espera se não houver relacionamento entre as variáveis ​​categóricas. No contexto do Qui-quadrado, a palavra "esperado" é equivalente ao que você esperaria se a hipótese nula fosse verdadeira. Se sua distribuição observada for suficientemente diferente da distribuição esperada (sem relacionamento), você pode rejeitar a hipótese nula e inferir que as variáveis ​​estão relacionadas.

Para um teste de Qui-quadrado, um valor de p menor ou igual ao seu nível de significância indica que há evidências suficientes para concluir que a distribuição observada não é a mesma que a distribuição esperada. Você pode concluir que existe uma relação entre as variáveis ​​categóricas.

- p >= 0.05 &rarr; Não podemos rejeitar a hipótese nula, portanto, não existe uma relação entre as variáveis ​​categóricas.
- p < 0.05 &rarr; Podemos rejeitar a hipótese nula e concluir que existe uma relação entre as variáveis ​​categóricas.

**Sources**: https://statisticsbyjim.com/hypothesis-testing/chi-square-test-independence-example/

Como na pergunta 2), vamos comparar as amostras contra a população. Portanto, vamos calcular o p, as médias e os desvios padrões de cada distribuição.
