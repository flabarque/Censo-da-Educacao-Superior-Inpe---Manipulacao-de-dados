### **SQL**: convertindo os arquivos *.csv em arquivos *.db 

```python
#fun√ß√£o para convertir arquivos *.csv em *.db
def convert_csv_to_sql(path_to_data, db_name, chunksize):
    '''
    Fun√ß√£o para convers√£o de arquivos csv em banco de dados sqlite3.
    
    INPUTS:
        path_to_data : str
            Caminho para os dados em formato csv a serem convertidos.
            
        db_name: str
            Nome do banco de dados a receber os dados
            
        chunksize : int
            Tamanho do lote de carregamento do arquivo csv.
    '''
    #criando conex√£o/banco de dados
    db = sqlite3.connect(db_name)

    #lendo todos os arquivos do caminho especificado em formato *.csv 
    files = glob.glob(f'{path_to_data}/*.CSV')
    
    #para cada um dos arquivos
    for file in files:

        #definindo o nome da tabela como sendo o nome do arquivo sem o *.csv
        #sim executando dum diretorio do PC: .split('\\')
        #sim executando dum diretorio do Google Drive: .split('/')
        table_name = file.split('/')[-1].split('.CSV')[0] 

        #carregando o arquivo indicado em chunks (lotes) de forma
        # a permitir opera√ß√£o com dataframes grandes demais
        df_chunks = pd.read_csv(file, sep='|', encoding='ISO-8859-1', chunksize=chunksize, low_memory=False)
        #alternativamente, usando encoding='latin1' obtem o mesmo resultado

        #para cada lote ("peda√ßo") do dataframe
        for df in df_chunks:

            #escrevendo o peda√ßo no banco
            df.to_sql(
                name=table_name, #definindo o nome da tabela a receber as informa√ß√µes (ou ser criada)
                index=False, #n√£o escreve o √≠ndice do dataframe na tabela
                if_exists='append', #se a tabela j√° existir, adiciona as informa√ß√µes abaixo das existentes
                con=db #definindo em qual banco os dados ser√£o inseridos
            )
```

A execu√ß√£o da fun√ß√£o tem uma dura√ß√£o de uns 16 minutos. V√°rios dos arquivos *.csv carregados s√£o grandes demais (e.g., > 2GB) com muitas linhas (e.g., 12043993 - ver abaixo).

```python
#utilizando a fun√ß√£o para convers√£o dos *.csvs em banco de dados sqlite
convert_csv_to_sql('/content/gdrive/My Drive/Colab Notebooks/dados', 'desafio_1.db', chunksize=100000)
```

Os valores 1, 2 e 3 s√£o Universidades P√∫blicas, e os valores 4 e 5 s√£o Universidades Privadas. Podemos utilizar as seguintes fun√ß√µes para agrupar essas categorias. 

```python
#criando categoria de Universidades P√∫blicas e Privadas a partir de 'TP_CATEGORIA_ADMINISTRATIVA'
def classificar_publica(categoria_administrativa):
    if categoria_administrativa < 3:
        return 'Universidade P√∫blica'  

def classificar_privada(categoria_administrativa):
    if categoria_administrativa > 3 < 5:
        return 'Universidade Privada'        

#agregando uma nova coluna 'CLF_PUBLICA_PRIVADA' com os valores trocados da coluna 'TP_CATEGORIA_ADMINISTRATIVA'
df_pub_pvd['CLF_PUBLICA'] = df_pub_pvd['TP_CATEGORIA_ADMINISTRATIVA'].apply(classificar_publica)
df_pub_pvd['CLF_PRIVADA'] = df_pub_pvd['TP_CATEGORIA_ADMINISTRATIVA'].apply(classificar_privada)

#tirando a coluna 'TP_CATEGORIA_ADMINISTRATIVA'
df_pub_pvd.drop('TP_CATEGORIA_ADMINISTRATIVA', axis='columns', inplace=True)
df_pub_pvd
```

##(b) 
Para resolver a quest√£o (b), vamos utilizar os dados das columnas `ID_ALUNO` (a.k.a., identifica√ß√£o dos estudantes), e `NO_IES` (a.k.a., nome Institui√ß√µes) e `CO_IES` (a.k.a., c√≥digo Institui√ß√µes) dos arquivos DM_ALUNO e DM_IES, respectivamente.

```python
# SQL

query = '''
SELECT t2.NO_IES, COUNT(t1.ID_ALUNO) AS NUM_ALUNOS
FROM DM_ALUNO t1
INNER JOIN DM_IES t2
ON t1.CO_IES = t2.CO_IES
GROUP BY t2.NO_IES
ORDER BY NUM_ALUNOS DESC
'''

df_estudantes_univ = pd.read_sql_query(query, db)
df_estudantes_univ
```

A presen√ßa de mulheres nos cursos de exatas pode ser descrita como uma distribui√ß√£o binomial. Vamos ent√£o calcular a m√©dia e o desvio padr√£o desta distribui√ß√£o. 

**Distribui√ß√£o Binomial:**

p &rarr; probabilidade de sucesso (probabilidade de ser mulher)

$$p = \frac{n_{mulheres}}{n_{total}} $$

$$ \mu = n_{total} * p $$

$$ \sigma = \sqrt{n_{total} * p * (1-p)} $$

**Sources**: https://pt.wikipedia.org/wiki/Distribui%C3%A7%C3%A3o_binomial

```python
#c√°lculo de p
n_mulheres = df_estudantes_exatas_pd.loc[df_estudantes_exatas_pd['TP_SEXO'] == 1, 'ID_ALUNO'].count()
n_total = df_estudantes_exatas_pd['ID_ALUNO'].count()
p = n_mulheres / n_total

#c√°lculo da m√©dia
mu = n_total*p

#c√°lculo do desvio padr√£o
std = np.sqrt(n_total * p * (1-p))
```

```python
print(f'n√∫mero de mulheres: {n_mulheres}','\n' f'n√∫mero total: {n_total}','\n' f'mu: {mu}','\n' f'desvio padr√£o: {std}')
```

Agora, vamos fazer uma **distribui√ß√£o normal** de uma popula√ß√£o hipot√©tica com a m√©dia $\mu$ e o desvio padr√£o $\sigma$ calculados para descrever a presen√ßa de mulheres nos cursos de exatas. Como queremos comparar a m√©dia de nossa amostra contra a m√©dia da popula√ß√£o hipot√©tica, podemos utilizar um teste t. Dessa maneira, podemos verificar se a m√©dia observada √© estatisticamente menor que a m√©dia esperada caso tivessem o mesmo n√∫mero de homens e mulheres nos cursos de exatas, ou seja, metade do n√∫mero de estudantes.

Vamos utilizar um teste t com grau de signific√¢ncia $\alpha$ de 0.05, com as seguintes hip√≥teses:

$$ H_0 : \mu_{observada} = \frac{n_{total}}{2} $$

$$ H_1 : \mu_{observada} < \frac{n_{total}}{2}$$

Vale perceber que estamos fazendo um teste monocaudal, testando se o n√∫mero de mulheres √© menor do que o n√∫mero de homens. Neste caso, o p-valor do teste monocaudal corresponde √† $p_{bicaudal} / 2 $. Al√©m disso, podemos olhar para o sinal do t-statistic para determinar se  m√©dia observada √© maior ou menor que a m√©dia esperada.
- t_statistic < 0 e p-valor < $\alpha$ &rarr; A m√©dia observada √© menor que a m√©dia esperada;
- t_statistic > 0 e p-valor < $\alpha$ &rarr; A m√©dia observada √© maior que a m√©dia esperada;
- p-valor > $\alpha$ &rarr; A m√©dia observada e a m√©dia esperada n√£o s√£o estatisticamente diferentes;

**Sources**: https://pt.wikipedia.org/wiki/Teste_t_de_Student

```python
#definindo grau de signific√¢ncia
alpha = 0.05
#criando a distribui√ß√£o normal hipot√©tica de m√©dia mu e desvio padr√£o std
norm_dist_estudantes_exatas = stats.norm.rvs(loc=mu, scale=std, size=n_total)

#fazendo o teste de hip√≥tese
t_statistic, p_value = stats.ttest_1samp(norm_dist_estudantes_exatas, int(n_total/2))

if p_value/2 >= alpha and t_statistic < 0:
    print(f'Com um p_valor de {round(p_value, 2)} n√£o rejeitamos a hip√≥tese nula, ou seja, n√£o existem menos mulheres do que homens nos cursos de exatas!')
else:
    print(f'Com um p_valor de {round(p_value, 2)} rejeitamos a hip√≥tese nula, ou seja, existem, de fato, menos mulheres do que homens nos cursos de exatas!')
```

A presen√ßa de mulheres nos cursos de exatas tamb√©m pode ser avaliada com um **teste de independ√™ncia do Qui-cuadrado**. Ele determina se existe uma rela√ß√£o estatisticamente significativa entre vari√°veis ‚Äã‚Äãcateg√≥ricas, nosso caso os g√™neros vs. os cursos de exatas. √â outro teste de hip√≥tese que responde √† pergunta - os valores de uma vari√°vel categ√≥rica dependem do valor de outras vari√°veis ‚Äã‚Äãcateg√≥ricas? Portanto:

**Hip√≥tese nula (H0)**: n√£o h√° rela√ß√µes entre as vari√°veis ‚Äã‚Äãcateg√≥ricas. Se voc√™ souber o valor de uma vari√°vel, isso n√£o o ajudar√° a prever o valor de outra vari√°vel. 

**Hip√≥tese alternativa (H1)**: Existem rela√ß√µes entre as vari√°veis ‚Äã‚Äãcateg√≥ricas. Saber o valor de uma vari√°vel ajuda a prever o valor de outra vari√°vel.

O teste de associa√ß√£o Qui-quadrado funciona comparando a distribui√ß√£o que voc√™ observa com a distribui√ß√£o que voc√™ espera se n√£o houver relacionamento entre as vari√°veis ‚Äã‚Äãcateg√≥ricas. No contexto do Qui-quadrado, a palavra "esperado" √© equivalente ao que voc√™ esperaria se a hip√≥tese nula fosse verdadeira. Se sua distribui√ß√£o observada for suficientemente diferente da distribui√ß√£o esperada (sem relacionamento), voc√™ pode rejeitar a hip√≥tese nula e inferir que as vari√°veis ‚Äã‚Äãest√£o relacionadas.

Para um teste de Qui-quadrado, um valor de p menor ou igual ao seu n√≠vel de signific√¢ncia indica que h√° evid√™ncias suficientes para concluir que a distribui√ß√£o observada n√£o √© a mesma que a distribui√ß√£o esperada. Voc√™ pode concluir que existe uma rela√ß√£o entre as vari√°veis ‚Äã‚Äãcateg√≥ricas.

- p >= 0.05 &rarr; N√£o podemos rejeitar a hip√≥tese nula, portanto, n√£o existe uma rela√ß√£o entre as vari√°veis ‚Äã‚Äãcateg√≥ricas.
- p < 0.05 &rarr; Podemos rejeitar a hip√≥tese nula e concluir que existe uma rela√ß√£o entre as vari√°veis ‚Äã‚Äãcateg√≥ricas.

**Sources**: https://statisticsbyjim.com/hypothesis-testing/chi-square-test-independence-example/


```python
from scipy.stats import chi2_contingency
from scipy.stats import norm
import math
```

Otras bibliotecas e fun√ß√µes:

*   **math**: este m√≥dulo fornece acesso √†s fun√ß√µes matem√°ticas definidas pelo padr√£o C (https://docs.python.org/3/library/math.html)
*   **chi2_contigencty**: esta fun√ß√£o calcula o Qui-quadrado e o valor p para o teste de hip√≥tese de independ√™ncia das frequ√™ncias observadas na tabela de conting√™ncia (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html)

*  **norm**: a normal continuous random variable. (https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html)

```python
#construindo a tabela de conting√™ncia
df_estudantes_exatas_pd_chi = pd.crosstab(index=df_estudantes_exatas_pd['TP_SEXO'], columns=df_estudantes_exatas_pd['NO_CINE_AREA_GERAL'], margins=True)
df_estudantes_exatas_pd_chi 
```

```python
#separando duas amostras, uma com alunos da regiao norte
# e outra com alunmos das demais regioes
CO_UF_regiao_norte = [11, 12, 13, 14, 15, 16, 17]

df_estudantes_regiao_norte = df_estudantes_regiao.loc[
    df_estudantes_regiao['CO_UF'].isin(CO_UF_regiao_norte)
]

df_estudantes_demais_regioes = df_estudantes_regiao.loc[
    ~df_estudantes_regiao['CO_UF'].isin(CO_UF_regiao_norte)
]

df_estudantes_regiao_norte.shape, df_estudantes_demais_regioes.shape
```

Como na pergunta 2), vamos comparar as amostras contra a popula√ß√£o. Portanto, vamos calcular o p, as m√©dias e os desvios padr√µes de cada distribui√ß√£o.

```python
#calculando estatisticas das distribui√ß√µes

#Regiao Norte
p_regiao_norte = df_estudantes_regiao_norte['DESISTENTE'].mean()
n_total_regiao_norte = df_estudantes_regiao_norte['ID_ALUNO'].count()
mu_regiao_norte = n_total_regiao_norte * p_regiao_norte
std_regiao_norte = np.sqrt(n_total_regiao_norte * p_regiao_norte * (1 - p_regiao_norte))

#Demais regioes
p_demais_regioes = df_estudantes_demais_regioes['DESISTENTE'].mean()
n_total_demais_regioes = df_estudantes_demais_regioes['ID_ALUNO'].count()
mu_demais_regioes = n_total_demais_regioes * p_demais_regioes
std_demais_regioes = np.sqrt(n_total_demais_regioes * p_demais_regioes * (1 - p_demais_regioes))
```

```python
#aproximando distribui√ß√µes para normais
norm_dist_regiao_norte = stats.norm.rvs(loc=mu_regiao_norte, scale=std_regiao_norte, size=n_total_regiao_norte)
norm_dist_demais_regioes = stats.norm.rvs(loc=mu_demais_regioes, scale=std_demais_regioes, size=n_total_demais_regioes)
```

```python
#fazendo teste de hip√≥tese para compara√ß√£o das duas distribui√ß√µes
t_statistic, p_value = stats.ttest_ind(norm_dist_regiao_norte, norm_dist_demais_regioes)

if t_statistic > 0 and p_value/2 < alpha:
    print(f'Como p_valor √© {round(p_value/2, 2)} e t_statistic √© positivo, rejeitamos a hip√≥tese nula e de fato existem mais desistentes na regi√£o Norte do que nas demais regi√µes')
elif t_statistic < 0 and p_value/2 < alpha:
    print(f'Como p_valor √© {round(p_value/2, 2)} e t_statistic √© negativo, rejeitamos a hip√≥tese nula e concluimos que existem menos desistentes na regi√£o Norte do que nas demais regi√µes')
else:
    print(f'Como p_valor √© {round(p_value/2, 2)} n√£o rejeitamos a hip√≥tese nula, ou seja, a propor√ß√£o de desistentes na regi√£o norte n√£o √© estatisticamente diferente das demais regi√µes')
```

Al√©m disso, como na pergunta 2), tamb√©m podem ser avaliados os estudantes desistentes com um teste de independ√™ncia do Qui-cuadrado. 

```python
#criando vari√°vel qualitativa que indica se o aluno √© da regi√£o norte (1) ou n√£o (0)
df_estudantes_regiao['NORTE'] = np.where(
    (df_estudantes_regiao['CO_UF'].isin(CO_UF_regiao_norte)),
    1,
    0
)
```

```python
#nomeando as categorias
df_estudantes_regiao['NORTE']= df_estudantes_regiao['NORTE'].map({0: 'N√£o', 1: 'Sim'})
df_estudantes_regiao['DESISTENTE']= df_estudantes_regiao['DESISTENTE'].map({0: 'N√£o', 1: 'Sim'})

#agrupando as propor√ß√µes
df_estudantes_regiao.groupby('NORTE')['DESISTENTE'].value_counts(normalize=True).to_frame()
```

```python
#construindo a tabela de conting√™ncia
df_estudantes_regiao_chi = pd.crosstab(index=df_estudantes_regiao['DESISTENTE'], columns=df_estudantes_regiao['NORTE'], margins=True)
df_estudantes_regiao_chi
```

```python
#ploteando o qui-cuadrado
df_estudantes_regiao_chi_plot = df_estudantes_regiao_chi.drop('All',axis=1)
df_estudantes_regiao_chi_plot = df_estudantes_regiao_chi_plot.drop('All',axis=0)
df_estudantes_regiao_chi_plot.plot(figsize=(12, 6), kind='bar', stacked=True, rot=0)
plt.title("Varia√ß√£o de desistencia dos estudantes nas diferentes regi√µes", fontsize=15, fontweight='bold')
plt.xlabel("Desist√™ncia", fontsize=15)
plt.ylabel("N√∫mero de estudantes (milh√µes)", fontsize=15)
plt.show()
```

```python
#calculando o qui-cuadrado e o valor de p
chi2, p, dof, ex = chi2_contingency(df_estudantes_regiao_chi, correction=False)

print(f'qui-cuadrado: {chi2}', '\n' f'valor de p: ' '{:.10f}'.format(p))

if p >= 0.05:
    print(f'Com um p_valor de {round(p, 2)} n√£o rejeitamos a hip√≥tese nula, ou seja, a propor√ß√£o de desistentes na regi√£o norte n√£o √© estatisticamente diferente das demais regi√µes!')
else:
    print(f'Com um p_valor de {round(p, 2)} rejeitamos a hip√≥tese nula, ou seja, a propor√ß√£o de desistentes na regi√£o norte √© estatisticamente diferente das demais regi√µes!')
```

Portanto, a propor√ß√£o de desist√™ncia dos estudantes da Regi√£o Norte √© estatisticamente menor (0.221878) que a propor√ß√£o de desist√™ncia dos estudantes do resto do pais (0.270400).

```python
#definindo novamente a vari√°vel desistente
df_estudantes_ies = df_estudantes_4.drop_duplicates(subset='ID_ALUNO').copy()
df_estudantes_ies['DESISTENTE'] = np.where(
    (df_estudantes_ies['TP_SITUACAO'] == 3) | (df_estudantes_ies['TP_SITUACAO'] == 4),
    1,
    0
)

#agrupando por IES para calcular a taxa de abandono por IES
df_estudantes_ies = df_estudantes_ies.groupby('CO_IES', as_index=False).agg(
    TAXA_ABANDONO = ('DESISTENTE', 'mean')
)
```

```python
#juntando os dois dataframes
df_estudante_docente = df_docente_ies.merge(df_estudantes_ies)
```

```python
#calculando correlac√µes:
pearson, p_pearson = stats.pearsonr(df_estudante_docente['ESCOLARIDADE_MEDIA'], df_estudante_docente['TAXA_ABANDONO'])
spearman, p_spearman = stats.spearmanr(df_estudante_docente['ESCOLARIDADE_MEDIA'], df_estudante_docente['TAXA_ABANDONO'])

print(f'Corral√ß√£o de Pearson: {round(pearson, 2)}, com p-valor de {round(p_pearson, 3)}')
print(f'Corral√ß√£o de Spearman: {round(spearman, 2)}, com p-valor de {round(p_spearman, 3)}')
```
